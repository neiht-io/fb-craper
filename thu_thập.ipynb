```python
FANPAGE_LINK ="phuclongcoffeeandtea"
FOLDER_PATH = "/content/drive/MyDrive/process"
COOKIE_PATH = "/content/drive/MyDrive/process/facebook.com_cookies.txt"

PAGES_NUMBER = 1
```

```python
from facebook_scraper import get_posts
import pandas as pd
```

```python
post_list = []
for post in get_posts(FANPAGE_LINK,
                    options={"comments": True, "reactions": True, "allow_extra_requests": True},
                    extra_info=True, pages=1, cookies=COOKIE_PATH):
    post_list.append(post)
```

#lấy danh sách url những người comment



```python
x= post_list[0]["comments_full"]

for comment in x:
    comment_id = comment['comment_id']
    commenter_name = comment['commenter_name']
    commenter_url = comment['commenter_url']
    print(f"{{'comment_id': '{comment_id}','commenter_name':'{commenter_name}','commenter_url':'{commenter_url}'}}")

```

```python
drive_path = '/content/drive/MyDrive/process'
```

```python


# Tạo và ghi file dưới dạng văn bản (text)
with open(drive_path + 'comments.txt', 'w') as file:
    for comment in x:
        comment_id = comment['comment_id']
        commenter_name = comment['commenter_name']
        commenter_url = comment['commenter_url']
        file.write(f"comment_id: {comment_id}, commenter_name: {commenter_name}, commenter_url: {commenter_url}\n")

# Tạo và ghi file dưới dạng CSV
import csv

with open(drive_path + 'comments.csv', 'w', newline='') as file:
    fieldnames = ['comment_id', 'commenter_name', 'commenter_url']
    writer = csv.DictWriter(file, fieldnames=fieldnames)

    # Viết header
    writer.writeheader()

    # Viết dữ liệu từ danh sách x
    for comment in x:
        comment_id = comment['comment_id']
        commenter_name = comment['commenter_name']
        commenter_url = comment['commenter_url']
        writer.writerow({'comment_id': comment_id, 'commenter_name': commenter_name, 'commenter_url': commenter_url})


print(f"Tệp văn bản đã được lưu tại: {drive_path}comments.txt")
print(f"Tệp CSV đã được lưu tại: {drive_path}comments.csv")

```

```python

post_df_full = pd.DataFrame(columns = [])

for post in post_list:
    post_entry = post
    fb_post_df = pd.DataFrame.from_dict(post_entry, orient='index')
    fb_post_df = fb_post_df.transpose()
    post_df_full = post_df_full.append(fb_post_df)
    print(post['post_id']+' get')
```
